```python
import time

url = f"https://api.data.gov.sg/v1/transport/traffic-images"
response = requests.get("https://api.data.gov.sg/v1/transport/traffic-images?")
traffic_url = response.json()

# Get the image URL for camera with ID 2701
woodlands_image_url = [i for i in traffic_url['items'][0]['cameras'] if i['camera_id'] == '2701'][0]['image']

# Download the image
image_response = requests.get(woodlands_image_url)

# Save the image to disk
with open("/Users/rachelp/Downloads/traffic_image_data_studio1/woodlands_image.jpg", "wb") as f:
    f.write(image_response.content)
    
    time.sleep(600)
```
